## Mushroom Classification Project

### Problem Description
The goal of this project is to classify mushrooms as **edible (e)** or **poisonous (p)** based on 20 descriptive features such as cap shape, cap color, stem characteristics, habitat, and season.  
This is a **binary classification problem** where the target variable is the mushroom class (`edible` or `poisonous`).  

---

### Project Workflow

#### 1. Exploratory Data Analysis (EDA)
- Inspect the dataset structure (rows, columns, data types).  
- Check for missing values and duplicates.  
- Visualize distributions of categorical features (bar plots) and numeric features (histograms, boxplots).  
- Explore correlations between features and the target class.  
- Look for class imbalance in the target variable (edible vs poisonous).  

#### 2. Data Cleaning & Preprocessing
- Handle missing values or unknown entries if present.  
- Encode categorical features (e.g., **One-Hot Encoding** or **Label Encoding**).  
- Scale numerical features (e.g., **StandardScaler** or **MinMaxScaler**).  
- Stratify the dataset when splitting into **training and test sets** to maintain class balance.  
- If class imbalance is significant, apply oversampling techniques (e.g., **SMOTE**).  

#### 3. Feature Selection
To reduce dimensionality and improve generalization, apply feature selection techniques such as:  
- **Filter Methods:** Use statistical tests (e.g., Chi-square, ANOVA F-test, mutual information) to rank features.  
- **Wrapper Methods:** Apply Recursive Feature Elimination (**RFE**) with models like Logistic Regression or Decision Trees.  
- **Embedded Methods:** Leverage feature importance from models (e.g., Decision Tree, Random Forest, Lasso Regression).  
- Compare model performance before and after feature selection to evaluate impact.  

#### 4. Model Training & Evaluation
Train and evaluate the following classification models:  
- **Logistic Regression**  
- **Decision Tree**  
- **Random Forest**  
- **Support Vector Machine (SVM)**  
- **K-Nearest Neighbors (KNN)**

For each model:  
- Train on the **training set**.  
- Evaluate on the **test set** using:  
  - **Classification Report** (Precision, Recall, F1-score, Accuracy)  
  - **Confusion Matrix**  
  - **Cross-validation (optional)** for stability check  

#### 5. Hyperparameter Tuning
To improve model performance, apply hyperparameter optimization techniques:  
- **Grid Search:** Exhaustively search all combinations of parameters (good for small parameter spaces).  
- **Random Search:** Randomly sample parameter combinations (more efficient for large parameter spaces).  
- **Bayesian Optimization:** Iteratively choose hyperparameters based on past performance (efficient for complex models like Random Forest, SVM, or KNN).  

Compare tuned models with baseline models to measure improvement.  

#### 6. Feature Importance
- Extract **feature importance** from Decision Tree and Random Forest models.  
- Rank features to identify the most influential ones for classification.  
- Use feature importance as a basis for feature selection and model simplification.  

#### 7. Model Comparison
- Compare performance across all models (baseline, tuned, and with feature selection).  
- Use metrics such as accuracy, precision, recall, F1-score, and ROC-AUC to decide the best-performing model.
- Create plot of ROC-AUC curve of each model with different color for comparison

# data cleaning and preprocessing

from the previous heatmap there are some columns with huge missing values so 
- i will drop stem-root , stem-surface , veil-type ,veil-color , spore-print-color
- i will try to impute the other columns [missing < 10k -> mode / missing > 10k -> i will use knn]

```python
# dropping features
X.drop(columns=['stem-root' , 'stem-surface' , 'veil-type' ,'veil-color' , 'spore-print-color'] , inplace=True)
```

```python
# imputing featuers with missing values < 10k
from sklearn.impute import SimpleImputer
SM = SimpleImputer(strategy='most_frequent')

X[['gill-attachment', 'ring-type']] = SM.fit_transform(X[['gill-attachment', 'ring-type']])
```

```python
# imputing features with missing values > 10k
from sklearn.preprocessing import OrdinalEncoder
from sklearn.impute import KNNImputer

encoder = OrdinalEncoder()
X[['cap-surface' , 'gill-spacing']] = encoder.fit_transform(X[['cap-surface' , 'gill-spacing']])

imputer = KNNImputer(n_neighbors=5)
X[['cap-surface' , 'gill-spacing']] = imputer.fit_transform(X[['cap-surface' , 'gill-spacing']]).round(0)
X[['cap-surface' , 'gill-spacing']] = encoder.inverse_transform(X[['cap-surface' , 'gill-spacing']])
```

```python
X.isna().sum()
```

```python
# visualize missing values
plt.figure(figsize = (15,5))
sns.heatmap(X.isna() , cmap='viridis' , cbar=False , yticklabels=False)
plt.show()
```

encoding categorical features
- i will use one hot encoding for season , has-ring , gill-spacing , does-bruise-or-bleed
- and i will use ordinal encoder for the other features

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Define categorical columns
cat_cols = ['season', 'has-ring', 'gill-spacing', 'does-bruise-or-bleed']

# ColumnTransformer with ONLY encoder
ct = ColumnTransformer(
    transformers=[
        ('encoder', OneHotEncoder(drop='first'), cat_cols)
    ],
    remainder='drop'
)

# Fit and transform
X_one_hot = ct.fit_transform(X[cat_cols])  # <--- use only categorical cols


# Get encoded feature names
ohe_feature_names = ct.named_transformers_['encoder'].get_feature_names_out(cat_cols)

# Convert to DataFrame
X_one_hot = pd.DataFrame(X_one_hot.toarray() if hasattr(X_one_hot, "toarray") else X_one_hot,
                               columns=ohe_feature_names,
                               index=X.index)

```

```python
X_one_hot
```

```python
# encode data using Bayesian encoding
from category_encoders import TargetEncoder
encoder = TargetEncoder(smoothing=2.0)

# Fit on training set
X[[
    'cap-shape' , 'cap-surface' , 'cap-color' , 'gill-attachment' ,
    'gill-color' ,  'stem-color' ,
    'ring-type' ,'habitat'
]] = encoder.fit_transform(X[[
    'cap-shape' , 'cap-surface' , 'cap-color' , 'gill-attachment' ,
    'gill-color' ,  'stem-color' ,
    'ring-type' ,'habitat'
]] , y)

X
```

```python
# compine the outputs from Bayesian encoder and one hot encoder
## droping repeated unencoded features
X.drop(columns=['season', 'has-ring', 'gill-spacing', 'does-bruise-or-bleed'] , inplace = True)

## combine the features
X = pd.concat([X , X_one_hot] , axis=1)


X
```

```python
# encoding the target column
OHE = OneHotEncoder(drop='first' , sparse_output=False)

y = OHE.fit_transform(y)
y
```

splitting the data into train and test

```python
# splitting the data
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(X , y , test_size=0.25 , stratify=y , random_state=42)
```

```python
# scaling the numerical features
## since all numerical features have right skew then i will use log1p
x_train['cap-diameter'] = np.log1p(x_train['cap-diameter'])
x_train['stem-height'] = np.log1p(x_train['stem-height']) 
x_train['stem-width'] = np.log1p(x_train['stem-width'])

x_test['cap-diameter'] = np.log1p(x_test['cap-diameter'])
x_test['stem-height'] = np.log1p(x_test['stem-height'])
x_test['stem-width'] = np.log1p(x_test['stem-width'])
```

```python
# check class balance
unique_vals, counts = np.unique(y_train, return_counts=True)

# Display results
for val, cnt in zip(unique_vals, counts):
    print(f"Value {val}: {cnt} times")
```

# feature selection

```python
# initilizing and fit random forest on the data
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(x_train , y_train)
```

```python
# getting features importance
importances = rf.feature_importances_
feature_importance_df = pd.DataFrame({
    'Feature': x_train.columns,
    'Importance': importances
}).sort_values(by="Importance", ascending=False)

plt.barh( y = feature_importance_df['Feature'] , width=feature_importance_df['Importance'])
plt.title('importance of each feature')
plt.xlabel('feature importane')
plt.ylabel('feature name')
plt.show()
```

```python
# select the highest importance feature
selector = SelectFromModel(rf, threshold="median", prefit=True)

x_train_selected = selector.transform(x_train)
x_test_selected = selector.transform(x_test)

print("Original shape:", x_train.shape)
print("Reduced shape:", x_train_selected.shape)
```

```python
rf2 = RandomForestClassifier(n_estimators=100, random_state=42)
rf2.fit(x_train_selected , y_train)
```

```python
# compare model before and after the feature selection
from sklearn.metrics import accuracy_score
before_selection_pred = rf.predict(x_test)
after_selection_pred = rf2.predict(x_test_selected)

print(f'the accuracy before feature selection = {accuracy_score(y_test ,before_selection_pred )}')
print(f'the accuracy after feature selection = {accuracy_score(y_test , after_selection_pred)}')
```

```python
predictions = [pred_tuned_lr , pred_tuned_dt ,pred_tuned_rf ,pred_tuned_sv , pred_tuned_knn ,pred_tuned_nb   ]
model_name = ['logistic regression' , 'decision tree' , 'random forest' , 'support vector' , 'knn' , 'naive bias']
plt.figure(figsize=(15 , 10))
fig = 1
plt.suptitle('confusion matrix for every tuned model')
for pred , name in zip(predictions , model_name) :
    plt.subplot(2,3,fig)
    conf = confusion_matrix(y_test , pred)
    sns.heatmap(conf , annot=True)
    plt.title(name)
    fig +=1
plt.tight_layout()
plt.show()
```

```python
# getting the roc values for each model
from sklearn.metrics import roc_curve , roc_auc_score
fbr_list , tpr_list = [] , []
for pred in predictions:
    fbr , tpr , _ = roc_curve(y_test , pred)
    fbr_list.append(fbr)
    tpr_list.append(tpr)
print(f'fbr values = {fbr_list}')
print(f'tpr values = {tpr_list}')
```

```python
# plotting each roc curve for each model
plt.figure(figsize=(15,10))
for fbr , tpr , name in zip(fbr_list , tpr_list , model_name):
    plt.plot(fbr , tpr , label = name)
plt.title('roc curves for tuned models')
plt.legend()
plt.show()
```

so the best performing model is random forest

```python
# getting auc values 
auc_values = []
for pred in predictions:
    auc_values.append(roc_auc_score(y_test , pred))

print(f'the auc list values = {auc_values}')
```

```python
#plot auc values
colors = plt.cm.viridis(np.linspace(0, 1, len(model_name))) 
plt.figure(figsize=(15,5))
plt.barh(width = auc_values , y = model_name , color =colors )
plt.title('auc values for each tuned model')
plt.xlabel('auc score')
plt.ylabel('model name')
plt.show()
```
